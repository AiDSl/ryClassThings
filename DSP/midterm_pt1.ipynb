```python
class AiDSlTrombone(thinkdsp.Chirp):
    def evaluate(self, ts):
        freqs = numpy.linspace(220,440, len(ts) - 1)
        print(freqs);
        return self._evaluate(ts, freqs)

    def _evaluate(self, ts, freqs):
        PI2=numpy.pi*2
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        ys = self.amp * numpy.cos(phases)
        return ys
        
AiDSlTrombone().make_wave().make_spectrum().plot();
```

```python
class AiDSlTrombone(thinkdsp.Chirp):
    def evaluate(self, ts):
        freqs = numpy.linspace(220,440, len(ts) - 1)
        return self._evaluate(ts, freqs)

    def _evaluate(self, ts, freqs):
        PI2=numpy.pi*2
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        ys = self.amp * numpy.cos(phases)
        return ys
        
AiDSlTrombone().make_wave().make_spectrum().plot();
```

```python
def getList(that,duration,framerate,start=0):
    n = round(duration * framerate)
    xts = start + numpy.arange(n) / framerate
    xys = that.evaluate(xts)
    return xys
print("x.max:",getList(x,1,10000).max());
print("x.mean:",getList(x,1,10000).mean());
print("x.std:",getList(x,1,10000).std());
```

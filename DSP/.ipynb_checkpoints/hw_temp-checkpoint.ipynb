```python
#if can`t load thinkdsp,run command:"pip install thinkx"
#it should work at least for a while
import thinkdsp;


import matplotlib.pyplot;

#function that should be sealed in matplotlib as a setting.
def preset():
        matplotlib.pyplot.grid();

#wave work implment
class using_wave():
    #member: filepath:string,wave:wave;
    def __init__(self,str):
        self.filepath    =str;
        self.wave        =thinkdsp.read_wave(self.filepath);
        self.sound       =self.wave.make_audio;
    def display(self):
        self.wave.plot();
        
preset();


a_wave=using_wave("376617__askeseverin__piano-and-violin-dramatic-sad-30-seconds.wav");
a_wave.display()
a_wave.sound()#if you type ";" then it would do nothing
```

```python
import lib_thinkdsp;
import thinkdsp;
import numpy;

import matplotlib.pyplot;


class AiDSl_Saw(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        phase=2*numpy.pi*self.freq;
        cut=self.offset+ts*phase
        cut%=2*numpy.pi;
        climb=((cut/numpy.pi)-1);
        return climb*self.amp;
```

```python
AiDSl_Saw(freq=1000,amp=10).plot();
matplotlib.pyplot.grid();
```

```python
AiDSl_Saw(freq=1000,amp=10).make_wave().make_spectrum().plot();
matplotlib.pyplot.grid();
```
